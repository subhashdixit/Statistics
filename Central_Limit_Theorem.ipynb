```python
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Histogram and Density plot of sample mean from size 1 to 40 to check the distribution variation**

```python
from numpy.random.mtrand import sample
population_size = 1000000
population = np.rand(1000000)

number_of_samples = 10000
sample_means = np.rand(number_of_samples)

# Keep changing sample size to check the distribution
sample_size_list = [1,2,5,10,20,30,40]
# sample_size = 1
c = np.rand(number_of_samples)

for sample_size in sample_size_list:
  for i in range(0, number_of_samples):
      c = np.randint(1, population_size, sample_size)
      sample_means[i] = population[c].mean()


  # Histogram
  plt.subplot(1, 2, 1)
  plt.xticks(fontsize = 14)
  plt.yticks(fontsize = 14)
  sns.distplot(sample_means, bins = int(180/5), hist = True, kde = False)
  plt.title('Histogram of Sample Means', fontsize = 20)
  plt.xlabel('Sample Mean', fontsize = 20)
  plt.ylabel('Count', fontsize = 20)

  # Smoothing of histogram using kde
  plt.subplot(1, 2, 2)
  plt.xticks(fontsize = 14)
  plt.yticks(fontsize = 14)
  sns.distplot(sample_means, hist = False, kde = True)
  plt.title('Density of Sample Means', fontsize = 20)
  plt.xlabel('Sample Mean', fontsize = 20)
  plt.ylabel('Count', fontsize = 20)
  plt.subplots_adjust(bottom = 0.1, right = 2, top = 0.9)
```

**As we can obsevre that more the number of size of the sample, graph will be more normally distrbuted**
